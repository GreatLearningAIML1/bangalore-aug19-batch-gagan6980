```python
Hr_external.columns.get_loc('Age')
a=Hr_external.iloc[:,2].quantile(.25)
b=Hr_external.iloc[:,2].quantile(.75)
IQR=b-a
print('75th percentile\t:\t%d\n25th percentile\t:\t%d\nIQR\t\t:\t%d'%(b,a,IQR))
```
