## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```python
# define function for label prediction
def yhat(x,m,c):
  y_predicted=tf.add(tf.matmul(x,m),c)
  return y_predicted
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```python
# define function for loss calculation
def loss_func(y_pred,y_act):
    error=tf.subtract(y_pred,y_act)
    loss=tf.reduce_mean(tf.square(error))
    return loss
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```python
# assign learing rate value
L=tf.constant(0.01)
```

```python
L
```

```python
# define  a function for optimization
def train_step(w,b,L,y,X_norm): 
    with tf.GradientTape(persistent=True) as tape:
      tape.watch([w,b]) 
      current_loss=loss_func(yhat(X_norm,w,b),y)
      # Calculate Gradients of Loss w.r.t weights and bia
      dw,db = tape.gradient(current_loss, [w,b])
      # Update Weights and Bias based on gradients
      w.assign_sub(L*dw)
      b.assign_sub(L*db)
     
     
   
   
    
```

## Question 10

### Train the model for 100 epochs 
- Observe the training loss at every iteration

```python
epochs=tf.constant(100)
# Train the model for 100 epochs 
for epoch_count in range(epochs):
    real_loss = loss_func(y, yhat(X_norm,w,b))
    train_step( w,b, L,y,X_norm)
    print(f"Epoch count {epoch_count}: Loss value: {real_loss.numpy()}")


```
